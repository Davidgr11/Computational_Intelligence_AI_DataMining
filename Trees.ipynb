```python
import xgboost
import pandas
import matplotlib.pyplot as pyplot
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import accuracy_score
```

```python
iris=load_iris() #guardas lo que importa en una variable y lo despliegas
iris
```

```python
X,y=iris.data,iris.target
X
```

```python
X[1:10]
```

```python
X_entrenamiento,X_examen,y_entrenamiento,y_examen=train_test_split(X,y,test_size=0.15)
#divide un grupo de entrenamiento y otro de examen, el de entrenamiento tiene más elementos
```

```python
arbol=tree.DecisionTreeClassifier()
```

```python
arbol.fit(X_entrenamiento,y_entrenamiento)
```

```python
tree.plot_tree(arbol)
```

```python
tree.plot_tree(arbol,feature_names=iris.feature_names,class_names=iris.target_names,filled=True)
```

```python
pred=arbol.predict(X_examen)#nos dice a que categoría pertenece esta flor
pred
```

```python
y_examen
```

```python
accuracy_score(y_examen,pred)#resultado
```

```python
arbol=xgboost.XGBClassifier(n_estimators=10)
arbol.fit(X_entrenamiento,y_entrenamiento)
```

```python
xgboost.plot_tree(arbol,num_trees=0)
```

```python
fig,axs=pyplot.subplots(3,3)
xgboost.plot_tree(arbol,num_trees=0,ax=axs[0,0])
xgboost.plot_tree(arbol,num_trees=0,ax=axs[0,1])
xgboost.plot_tree(arbol,num_trees=0,ax=axs[0,2])
xgboost.plot_tree(arbol,num_trees=0,ax=axs[1,0])
xgboost.plot_tree(arbol,num_trees=0,ax=axs[1,1])
xgboost.plot_tree(arbol,num_trees=0,ax=axs[1,2])
xgboost.plot_tree(arbol,num_trees=0,ax=axs[2,0])
xgboost.plot_tree(arbol,num_trees=0,ax=axs[2,1])
xgboost.plot_tree(arbol,num_trees=0,ax=axs[2,2])
pyplot.show()
```
