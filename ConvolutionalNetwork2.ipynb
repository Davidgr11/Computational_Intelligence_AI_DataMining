```python
from matplotlib import pyplot
import numpy

from tensorflow.keras.datasets.cifar10 import load_data
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Conv2D,MaxPool2D
from tensorflow.keras.layers import Flatten, Dropout
```

```python
(X,y),(examenX,exameny)=load_data()
```

```python
X.shape
```

```python
examenX.shape
```

```python
for i in range(9):
  pyplot.subplot(3,3,i+1)
  pyplot.imshow(X[i],cmap=pyplot.get_cmap('gray'))
  pyplot.title(label='Imagen {}={}'.format(i,y[i]))
pyplot.tight_layout()
pyplot.show()
```

```python
X.shape
```

```python
categorias=len(numpy.unique(y))
categorias
```

```python
numpy.max(X[0])
```

```python
X=X.astype('float32')/255
```

```python
numpy.max(X[0])
```

```python
numpy.max(examenX[0])
```

```python
examenX=examenX.astype('float32')/255
numpy.max(examenX[0])
```

```python
numpy.max(examenX[0])
```

```python
red=Sequential()
red.add(Conv2D(32,(3,3),activation='relu',
               input_shape=X.shape[1:]))
red.add(MaxPool2D((2,2)))
red.add(Flatten())
red.add(Dense(100,activation='relu'))
red.add(Dropout(0.5))
red.add(Dense(categorias,activation='softmax'))
red.compile(optimizer='adam',
            loss='sparse_categorical_crossentropy',
            metrics=['accuracy'])
```

```python
red.fit(X,y,epochs=15,batch_size=2)
```
