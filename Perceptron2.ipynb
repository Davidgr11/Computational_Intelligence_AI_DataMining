```python
import pandas
import numpy
from matplotlib import pyplot
from sklearn.datasets import load_iris
from sklearn.linear_model import Perceptron
```

```python
iris=load_iris()
```

```python
datos=pandas.DataFrame(iris.data,
                       columns=iris.feature_names)
target=pandas.DataFrame(iris.target,
                        columns=['target'])
```

```python
datos
```

```python
target
```

```python
pyplot.scatter(datos.iloc[:,1],
               datos.iloc[:,2],
               c=target['target']
               )
pyplot.xlabel(datos.iloc[:,1].name)
pyplot.ylabel(datos.iloc[:,2].name)
pyplot.grid(True)
pyplot.show()
```

```python
datos=datos[target['target']!=2]
datos
```

```python
target=target[target['target']!=2]
```

```python
pyplot.scatter(datos.iloc[:,1],
               datos.iloc[:,2],
               c=target['target']
               )
pyplot.xlabel(datos.iloc[:,1].name)
pyplot.ylabel(datos.iloc[:,2].name)
pyplot.grid(True)
pyplot.show()
```

```python
red=Perceptron(max_iter=100,tol=1e-3,shuffle=True)
```

```python
red.fit(datos.iloc[:,1:3],target)
```

```python
red.coef_
```

```python
red.intercept_
```

```python
red.score(datos.iloc[:,1:3],target)
```

```python
pesos=red.coef_[0]
pesos
```

```python
pendiente=-1*(pesos[0]/pesos[1])
ordenada=-1*red.intercept_/pesos[1]
```

```python
fig,ax=pyplot.subplots()
pyplot.scatter(datos.iloc[:,1],
               datos.iloc[:,2],
               c=target['target']
               )
xvals=numpy.array(ax.get_xlim())
yvals=ordenada+pendiente*xvals
pyplot.plot(xvals,yvals,'--')
pyplot.title('Perceptron')
pyplot.show()
```
